# scatter plot

```python
n=300
x=np.random.rand(n)
y=np.random.rand(n)
colors=np.random.rand(n)
area=np.pi*(20*np.random.rand(n)**2)
```

```python
plt.figure(figsize=(14,6))
plt.scatter(x,y,c=colors,s=area,cmap='Spectral')
plt.colorbar()
plt.show()
```

```python
fig=plt.figure(figsize=(14,6))
ax1=fig.add_subplot(1,2,1)
plt.scatter(x,y,s=area,c=colors,cmap="Pastel1")
plt.colorbar()

ax2=fig.add_subplot(1,2,2)
plt.scatter(x,y,s=area,c=colors,cmap='Pastel2')
plt.colorbar()

plt.show()
```

# KDE (kernel density estimation)

```python
from scipy import stats
density=stats.kde.gaussian_kde(values)

plt.subplots(figsize=(12,6))

values2=np.linspace(min(values)-10,max(values)+10,100)
plt.plot(values2,density(values2))
plt.fill_between(values2,0,density(values2),color='pink')

plt.show()
```

# combine plot

```python
plt.subplots(figsize=(12,6))

plt.hist(values,bins=100,density=1,histtype='bar',color='grey',linewidth=5)
plt.plot(values2,density(values2))
plt.xlim(xmin=-5,xmax=5)
plt.show()
```
